```python
data = {
    'Name': ['Aryan','Sakshi','Sonal','Usha'],
    'Marks':[99,95,70,100]
}

df = pd.DataFrame(data)
```

```python
df2 = {
    'NewName' : ['Jayshree','Muskan','Pranav','Rahul'],
    'Marks':[99,95,70,100],
    'Age': [25,26,27,28]
}
```

```python
df2 = pd.DataFrame(df2)
```

```python
df2
```

```python
df
```

```python
pd.merge(df,df2,on='Marks') # SQL Join
```

```python
pd.concat([df,df2]) # vertical
```

```python
pd.concat([df,df2],axis=1)
```
